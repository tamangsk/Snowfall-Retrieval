```
device = torch.device("cuda:0" if torch.cuda.is_available() else 'cpu')
```

```
# a)setting up neural net
class neural_net(nn.Module):
  def __init__(self):
      super(neural_net, self).__init__()
      self.fc1 = nn.Linear(21,256) # 26 feature
      self.fc2 = nn.Linear(256,512)
      self.fc3 = nn.Linear(512, 256)
      self.fc4 = nn.Linear(256,128)
      self.fc5 = nn.Linear(128,1)
      self.fc6 = nn.Linear(2,32)
      self.fc7_mu = nn.Linear(32, 8) # output of the mu (mean)
      self.fc7_sigma = nn.Linear(32, 8)
      self.relu = nn.ReLU()
  def encode(self, x):
      x = self.relu(self.fc6(x))
      mu = self.fc7_mu(x)
      sigma = self.fc7_sigma(x)
      return mu, sigma

  def reparameterize(self, mu, sigma):
      std = torch.exp(0.5*sigma) # standard deviation ( sigma = log((std)**2) )
      noise = torch.randn_like(std)
      z = mu + std*noise # The "source" to the decoder and we want to make this to unit Gaussian after training.
      return z 
  def predictphase(self, x):
      z= torch.normal(0, 1, size=(x.shape[0], 8)).to(device)
      #z=torch.randn(x.shape[0], 8).to(device)
      x= x[:,0:13]
      x= torch.cat((x,z),1)
      x = self.fc1(x) #
      x = self.relu(x) #25         
      x = torch.flatten(x, start_dim=1)
      x = self.relu(self.fc2(x))
      x = self.relu(self.fc3(x))
      x = self.relu(self.fc4(x))
      x = self.fc5(x)
      return x
  def forward(self, x):
      mu, sigma = self.encode(x[:,13:15])
      z = self.reparameterize(mu, sigma)
      x= x[:,0:13]
      #print(z.shape,x.shape)
      #z= self.encode(x[-1])
      #print(z.shape())
      #x= 
      x= torch.cat((x,z),1)
      #print(x.shape)
      x = self.fc1(x) #
      x = self.relu(x) #25   
      #x = self.droput(x)        
      x = torch.flatten(x, start_dim=1)
      x = self.relu(self.fc2(x))
      x = self.relu(self.fc3(x))
      x = self.relu(self.fc4(x))
      #x = self.droput(x) 
      x = self.fc5(x)

      return x,mu,sigma
```

```
def loss_function(output, real, mu, sigma):
    loss = nn.MSELoss()
    MSE = loss(output, real) # MSE loss
    kl_divergence = -0.5*torch.sum(1 + sigma - mu.pow(2) - sigma.exp()) # KL-divergence term

    return MSE + kl_divergence, MSE
```

```
def fit(model,trainloader):
  Loss = nn.MSELoss()
  optimizer = optim.AdamW(model.parameters(), lr)
  testerror =[]
  trainerror =[]
  #d-dimensional vector representing diag(Gt) to store a running total of the squares of the gradients.
  for e in range(epochs):
      running_loss = 0
      for input in trainloader:  
          input = input.to(device)    
          optimizer.zero_grad()         
          output,mu,sigma = model(input[:,0:15].float())
          #print(output)
          #output = output.view(-1,1,1,1)
          loss,mse = loss_function(output,input[:,15].float().view(-1,1),mu,sigma)
          #loss = Loss(output, input[:,15].float().view(-1,1))
          #print(loss)
          loss.backward()
          optimizer.step()
          running_loss += mse.item()
      # if(e!=0 and e!=1):     
      #   testerror.append(evaluate_model(neural,test_loader))
      #   trainerror.append(math.sqrt(running_loss/len(trainloader)))        
      print(e,math.sqrt(running_loss/len(trainloader)))
      #print(evaluate_model(neural,test_loader))
      if(math.sqrt(running_loss/len(trainloader))<0.04):
        break
          #print("Epoch {} - Training loss: {} -Test Accuracy:{}".format(e, running_loss))
  # epochs_list = range(2,epochs)        
  # plt.plot(epochs_list,trainerror,'g', label='Training loss')
  # plt.plot(epochs_list,testerror,'r', label='Test loss')
  # plt.title('Training and TEST RMSE loss')
  # plt.ylabel('RMSE Error')
  # plt.xlabel('Epochs')
  # plt.legend()
  # plt.grid(True)
  # plt.show()           
```

```
from sklearn.metrics import mean_squared_error
```

```
def evaluate_model(model, test_loader):
    predictions, actuals = list(), list()
    model.eval()
    for inputs in test_loader:
        # evaluate the model on the test set
        inputs = inputs.to(device)
        yhat = model.predictphase(inputs[:,0:15].float())
        # retrieve numpy array
        yhat = yhat.cpu().detach().numpy()
        actual = inputs[:,15].cpu().float().numpy()
        actual = actual.reshape((len(actual), 1))
        # store
        predictions.append(yhat)
        actuals.append(actual)
    predictions, actuals = np.vstack(predictions), np.vstack(actuals)
    # calculate mse
    mse = mean_squared_error(actuals, predictions)
    rmse = math.sqrt(mse)
    return rmse
```
